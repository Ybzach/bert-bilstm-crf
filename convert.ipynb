```python
def convert_to_bio_with_pos(text, annotations):
    doc = nlp(text)
    tokens = [(token.text, 'O') for token in doc]  # Initialize tokens with 'O' label

    for annotation in annotations:
        start = annotation["start"]
        end = annotation["end"]
        labels = annotation["labels"]
        label = labels[0] if labels else "O"

        # Convert character indices to token indices
        char_span = doc.char_span(start, end)
        if char_span is None:
            continue
        
        start_token = char_span.start
        end_token = char_span.end

        # Adjust token labels
        for i in range(start_token, end_token):
            if i == start_token:
                tokens[i] = (tokens[i][0], 'B-' + label)
            else:
                tokens[i] = (tokens[i][0], 'I-' + label)
    
    return tokens
```
